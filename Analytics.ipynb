```python
upper_limit = df.TotalWorkingYears.mean() + 3*df.TotalWorkingYears.std()
lower_limit = df.TotalWorkingYears.mean() - 3*df.TotalWorkingYears.std()
df_no_outlier = df[(df.TotalWorkingYears < upper_limit) & (df.TotalWorkingYears > lower_limit)]
df_no_outlier.shape
```

```python
upper_limit = df.YearsAtCompany.mean() + 3*df.YearsAtCompany.std()
lower_limit = df.YearsAtCompany.mean() - 3*df.YearsAtCompany.std()
df_no_outlier = df_no_outlier[(df_no_outlier.YearsAtCompany < upper_limit) & (df_no_outlier.YearsAtCompany > lower_limit)]
df_no_outlier.shape
```

```python
upper_limit = df.YearsInCurrentRole.mean() + 3*df.YearsInCurrentRole.std()
lower_limit = df.YearsInCurrentRole.mean() - 3*df.YearsInCurrentRole.std()
df_no_outlier = df_no_outlier[(df_no_outlier.YearsInCurrentRole < upper_limit) & (df_no_outlier.YearsInCurrentRole > lower_limit)]
df_no_outlier.shape
```

```python
upper_limit = df.YearsSinceLastPromotion.mean() + 3*df.YearsSinceLastPromotion.std()
lower_limit = df.YearsSinceLastPromotion.mean() - 3*df.YearsSinceLastPromotion.std()
df_no_outlier = df_no_outlier[(df_no_outlier.YearsSinceLastPromotion < upper_limit) & (df_no_outlier.YearsSinceLastPromotion > lower_limit)]
df_no_outlier.shape
```

```python
upper_limit = df.YearsWithCurrManager.mean() + 3*df.YearsWithCurrManager.std()
lower_limit = df.YearsWithCurrManager.mean() - 3*df.YearsWithCurrManager.std()
df_no_outlier = df_no_outlier[(df_no_outlier.YearsWithCurrManager < upper_limit) & (df_no_outlier.YearsWithCurrManager > lower_limit)]
df_no_outlier.shape
```
